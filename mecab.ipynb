```python
file_path = "/Users/nakazawamika/Documents/研究室/オフィス/データ/5-18_normal.txt" 

#ファイルを開く　with：ファイルを自動的に閉じてくれる　r：読み込み
with open(file_path, "r", encoding="utf-8", errors="replace") as file:  
    text = file.read()
```

```python
#辞書指定
tagger = MeCab.Tagger()
#()の中変えると出力のフォーマットが変わる
```

```python
# 名詞のみを抽出する関数
def extract_nouns(text):
    nouns = []  #名詞を格納するリスト
    node = tagger.parse(text)   #parse 形態素解析
    lines = node.split('\n')
    for line in lines:
        if '名詞' in line:
            noun = line.split('\t')[0]  #\t タブで分割 0番目の要素＝単語を取り出す
            nouns.append(noun)
    return nouns

node = tagger.parse(text)
print(node)
```
