```python
# 形態素解析した結果をリスト化(名詞のみ)

# def extract(train):
#     """Morphological analysis of text and returning a list of only nouns"""
#     tagger = MeCab.Tagger('-Ochasen')
#     node = tagger.parseToNode(train)
#     keywords = []
#     while node:
#         if node.feature.split(",")[0] == u"名詞": # this means noun
#             keywords.append(node.surface)
#         node = node.next
#     return keywords

# Taggerオブジェクトを生成
tokenizer = MeCab.Tagger("")
tokenizer.parse("")

def extract(text):
    words = []

    # 単語の特徴リストを生成
    node = tokenizer.parseToNode(text)

    while node:
        # 品詞情報(node.feature)が名詞ならば
        if node.feature.split(",")[0] == u"名詞":
            # 単語(node.surface)をwordsに追加
            words.append(node.surface)
        node = node.next

    # 半角スペース区切りで文字列を結合
    text_result = ' '.join(words)
    return text_result
```

```python
docs = []

for i in file_contents:
    text = extract(i)
    docs.append(text)

print(docs)
```

```python
# モデルを生成
vectorizer = TfidfVectorizer(smooth_idf=False)
X = vectorizer.fit_transform(docs)

# データフレームに表現
values = X.toarray()
feature_names = vectorizer.get_feature_names_out()
df = pd.DataFrame(values, columns = feature_names,
                  index=["吾輩は猫である", "三四郎", "こころ"])
print(df)
```
